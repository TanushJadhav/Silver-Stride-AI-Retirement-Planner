```python
#Importing the required Libraries

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("Retired_dataset.csv")
df
```

```python
#Diving the Spending Scores in to # categoriesL Low, Average and High

df['Spending Score'] = df['Spending Score'].replace([range(0,30), range(30,60), range(60,101)], ['low', 'average', 'high'])
```

```python
df
```

Model Building

```python
#ML algorithms cannot handle objects, so we are mapping the objects to integers values

df['Profession'] = df['Profession'].replace(['Artist','Entertainment', 'Lawyer', 'Engineer', 'Executive',
'Marketing', 'Homemaker', 'Healthcare'], [1,2,3,4,5,6,7,8])

df['Marital Status'] = df['Marital Status'].replace(['Married', 'Single','Widowed','Divorced'], [1,2,3,4])

df['Spending Score'] = df['Spending Score'].replace(['low', 'average', 'high'], [0,1,2])
```

```python
df
```

```python
x = df.drop('Spending Score', axis=1)
y = df['Spending Score']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestClassifier()

rf.fit(x_train, y_train)
print(classification_report(y_test, rf.predict(x_test)))
```

```python
rf.predict([[75,4,1,180000,70000]])
```
